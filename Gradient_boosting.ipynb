```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[165,137,472,192],[101,92,250,144],[29,127,201,91]], columns= ['R&D','ops','Marketing','profit'])
df
```

```python
df['f0'] = df['profit'].mean() # f0 i y_mean is model 1
```

```python
df
```

```python
# pseudo residual : f0 - profit
df['r1'] = df['f0'] - df['profit']  # r1 is for the model 2
df
```

```python
from sklearn.tree import DecisionTreeRegressor

reg = DecisionTreeRegressor(max_depth = 3 )  # Dt for model 2 by taking output as r1
reg.fit(df.iloc[:,0:3],df.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
plot_tree(reg)
```
