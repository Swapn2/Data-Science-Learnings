```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/imdb-top-1000.csv')
```

```python
movies.head(5)
```

```python
genre = movies.groupby('Genre')
```

```python
genre.sum()
```

```python
genre.min()
```

```python
# find the top 3 genre by total earning
genre.sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
genre['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
#find the genre with highest avg imdb rating
genre['IMDB_Rating'].mean().sort_values(ascending=False).head(3)
```

```python
# find the director with most popularity
director = movies.groupby('Director')
```

```python
director['No_of_Votes'].sum().sort_values(ascending=False).head(3)
```

```python
# find the number of done by each actor

movies['Star1'].value_counts()
```

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```
