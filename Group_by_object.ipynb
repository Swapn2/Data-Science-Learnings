```python
# agg method
# passing dict

genre.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
genre.agg(['min','max'])
```

```python
# Adding both the syntax
genre.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# find the highest rated movie in each genre
# looping on groups
df = pd.DataFrame(columns=movies.columns)
for group,data in genre:
  df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]])
```

```python
# split apply combine
genre.apply(min)
```

```python
# find number of movies starts with A each group
def foo(group_genre):
  return group_genre['Series_Title'].str.startswith('A').sum()
```

```python
genre.apply(foo)
```

```python
 # find ranking of each movie in the group according to the IMDB score
def rank_movie(genre):
   genre['genre_rank']  = genre['IMDB_Rating'].rank(ascending = True)
   return genre
```

```python
genre.apply(rank_movie)
```
