```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

n_samples = 300
n_features = 2
centres = 4

x,y = make_blobs(n_samples = n_samples , n_features = n_features, centers = centres , cluster_std = 1 , random_state = 42 )

plt.scatter(x[:,0],x[:,1], s = 50 , cmap = 'viridis')
plt.title("Genrated Datasets for K - means clustering ")
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_init = 10, n_clusters= i)
  kmeans.fit(x,y)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_init = 10, n_clusters= 4)
```

```python
y_means = kmeans.fit_predict(x)
```

```python
y_means
```

```python
plt.scatter(x[y_means == 0 , 0],x[y_means == 0 , 1], s = 60 , c = 'red' , label = 'cluster1')
plt.scatter(x[y_means == 1 , 0],x[y_means == 1 , 1], s = 60 , c = 'blue' , label = 'cluster2')
plt.scatter(x[y_means == 2 , 0],x[y_means == 2 , 1], s = 60 , c = 'green' , label = 'cluster3')
plt.scatter(x[y_means == 3 , 0],x[y_means == 3 , 1], s = 60 , c = 'yellow' , label = 'cluster4')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 100 , c = 'black', label = 'centroids')

plt.legend()
plt.show()
```
