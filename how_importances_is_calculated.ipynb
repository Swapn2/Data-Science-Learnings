```python
from sklearn.datasets import make_classification
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y = make_classification(n_samples = 5 ,n_classes = 2 , n_features = 2 , n_informative = 2 , n_redundant=0,
                          random_state = 0)
```

```python
 x,y = make_classification(n_samples = 15 ,n_classes = 2 , n_features = 2 , n_informative = 2 , n_redundant=0,
                          random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 2)
rf.fit(x,y)
```

```python
rf.feature_importances_
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)
```
