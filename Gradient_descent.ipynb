```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x ,y = make_regression(n_samples = 4 , n_features= 1 , noise = 80, random_state = 13)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# lets apply gradient descent by taking the m value as constant = 78.35
# and start b from b = 0
y_pred = ((78.35)*x +0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label = 'ols')
plt.plot(x,y_pred,color = 'blue',label = 'b=0')
plt.legend()
# plt.show()
```

```python
b = 0
m = 78.35
b_slope = -2*np.sum(y-(m*x.ravel())-b)
b_slope
```

```python
lr = .1
step_size = lr*b_slope
```

```python
b = b- step_size
b
```

```python
y_pred1 = ((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred,color = 'blue',label = 'b=0')
plt.plot(x,y_pred1,color = 'green',label = 'b_new')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2*np.sum(y - (m*x.ravel())-b)
loss_slope

step_size = lr*loss_slope
step_size

b = b-step_size
b
print(b)

y_pred2 = ((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred,color = 'blue',label = 'b=0')
plt.plot(x,y_pred1,color = 'green',label = 'b_new')
plt.plot(x,y_pred2,color = 'yellow',label = 'b_new2')
plt.legend()
plt.show()
```

```python
# iteration 3

# iteration 2
loss_slope = -2*np.sum(y - (m*x.ravel())-b)
loss_slope

step_size = lr*loss_slope
step_size

b = b-step_size
b
print(b)

y_pred3 = ((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred,color = 'blue',label = 'b=0')
plt.plot(x,y_pred1,color = 'green',label = 'b_new')
plt.plot(x,y_pred2,color = 'yellow',label = 'b_new2')
plt.plot(x,y_pred3,color = 'purple',label = 'b_new3')
plt.legend()
plt.show()
```

```python
b = 0
m = 78.35
lr = .1
epoch = 10
for i in range(epoch):
  loss_slope = -2*np.sum(y-(m*x.ravel())-b)
  step_size = lr*loss_slope
  b = b - step_size
  y_pred = ((78.35*x)+b).reshape(4)
  plt.plot(x,y_pred)

plt.scatter(x,y)
```
