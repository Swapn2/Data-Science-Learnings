```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_regression(n_samples = 100,n_features = 1 , n_informative = 1 , n_targets = 1 , noise= 20 , random_state =13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
r = Ridge(alpha = 10 )
r.fit(x,y)
print(r.coef_)
print(r.intercept_)
```

```python
r1 = Ridge(alpha =100)
r1.fit(x,y)
print(r1.coef_)
print(r1.intercept_)
```

```python
plt.plot(x,y,'b.')
plt.plot(x,lr.predict(x),color = "green",label = "alpha = 0")
plt.plot(x,r.predict(x),color = "orange",label = "alpha = 10")
plt.plot(x,r1.predict(x),color = "red",label = "alpha = 100")
```
