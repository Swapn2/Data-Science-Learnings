```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/play_tennis.csv')
df
```

```python
pd.crosstab(df['outlook'],df['play'],normalize = 'all', margins = True)
```

```python
pd.crosstab(df['outlook'],df['play'],normalize = 'columns', margins = True) # it is giving the conditional probability
```

```python
d = {}
for i in ['outlook','temp','humidity','wind']:
  d.update(pd.crosstab(df[i],df['play'],normalize = 'columns', margins = True).stack().to_dict())
```
