```python
from sklearn.datasets import make_classification
import numpy as np
x,y = make_classification(n_samples = 100, n_features = 2, n_informative = 1, n_classes = 2,n_clusters_per_class = 1 , random_state = 41, n_redundant = 0, hypercube = False, class_sep = 20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c = y, cmap= 'winter',s = 100)
```

```python
def sigmoid(z):
  return 1/(1+ np.exp(-z))
```

```python
def perceptron(x,y):
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = .1
  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = sigmoid(np.dot(x[j],weights))
    weights = weights + lr*(y[j]- y_hat)*x[j]

  return weights[0],weights[1:]
```

```python
intercept,coef = perceptron(x,y)
```

```python
intercept
```

```python
coef
```

```python
m = -(coef[0]/coef[1])
b = - (intercept/coef[1])
```

```python
m
```

```python
b
```

```python
x_input = np.linspace(-3,3,100)
y_output = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_output,color = 'red',linewidth = 3)
plt.scatter(x[:,0],x[:,1],c = y, cmap= 'winter',s = 100)
plt.ylim(-3,2)
```
